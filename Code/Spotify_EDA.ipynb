```python
playlist_ids = {"37i9dQZF1DX4DZAVUAwHMT": "Rock", 
                "37i9dQZF1DXbHcQpOiXk1D": "Jazz", 
                "37i9dQZF1DX1kCIzMYtzum":"EDM",
                "37i9dQZF1DX4arVIN5Cg4U":"Dubstep",
                "37i9dQZF1EIeKh45OZ1ylm":"Techno",
                "37i9dQZF1DX2P4yz1JwPAz":"R&B",
                "37i9dQZF1EQmPV0vrce2QZ":"Country",
                "37i9dQZF1DWSVtp02hITpN":"Pop",
                "37i9dQZF1DX5uokaTN4FTR":"Electro",
                "37i9dQZF1DX2sUQwD7tbmL":"Indie Rock"}
```

```python
playlists = []
for idd in playlist_ids.keys():
    track_ids = get_playlist_tracks(spotify, idd)[:50]
    playlists.append(get_audio_features_ids(spotify, track_ids))
```

```python
fig, axs = plt.subplots(2,4,figsize=(20,10))
column = playlists[0].columns

for i in range(len(column)):
    row = int(i/4)
    col = int(i%4)
    counter = 0
    axs[row][col].set_title(column[i])
    for playlist in playlists:
        y = playlist[column[i]].sort_values()
        axs[row][col].scatter(range(len(playlist)), y, color = sns.color_palette()[counter], label = playlist_ids[list(playlist_ids.keys())[counter]], alpha = 0.4)
        counter += 1
plt.legend()
plt.savefig("../../figures/spotify_eda.png")
plt.show()
```
